#Скалярний добуток для визначення кута між векторами

Для перевірки результату можна використати калькулятор

https://onlinemschool.com/math/assistance/vector/angl/


```python
import numpy as np
import matplotlib.pyplot as plt

# Координати першого вектора
v1 = np.array([2, 3])

# Координати другого вектора
v2 = np.array([-1, 5])

# Створення графіку
plt.figure()

# Зображення першого вектора
plt.quiver(0, 0, v1[0], v1[1], angles='xy', scale_units='xy', scale=1, color='b', label='Вектор v1')

# Зображення другого вектора
plt.quiver(0, 0, v2[0], v2[1], angles='xy', scale_units='xy', scale=1, color='r', label='Вектор v2')

# Налаштування осей
plt.xlim(-2, 3)
plt.ylim(-1, 6)
plt.xlabel('X')
plt.ylabel('Y')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)

# Відображення легенди
plt.legend()

# Показ графіка
plt.show()
```

```python
# Обчислення скалярного добутку
#dot_product = np.dot(v1, v2)
dot_product = v1 @ v2
print('dot_product', dot_product)

# Обчислення довжин векторів
length_v1 = np.linalg.norm(v1)
length_v2 = np.linalg.norm(v2)
print('length_v1', length_v1)
print('length_v2', length_v2)

# Обчислення косинуса кута між векторами
cos_theta = dot_product / (length_v1 * length_v2)
print('cos_theta', cos_theta)

# Обчислення кута в радіанах
theta_radians = np.arccos(cos_theta)
print('theta_radians', theta_radians)

# Переведення кута в градуси
theta_degrees = np.degrees(theta_radians)
print('theta_degrees', theta_degrees)

print("Кут між векторами:", theta_degrees, "градусів")
```

#Векторний добуток для визначення нормалі до площини



```python
import numpy as np

def normal_vector(u, v):
    # Обчислення векторного добутку (cross product)
    n = np.cross(u, v)

    return n

# Приклад використання
u = np.array([1,-4, 0])
v = np.array([1, 2, 0])

normal = normal_vector(u, v)
print("Вектор нормалі до площини:", normal)
```

```python
# Створення графіку
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Зображення векторів
ax.quiver(0, 0, 0, u[0], u[1], u[2], color='r', label='Вектор u')
ax.quiver(0, 0, 0, v[0], v[1], v[2], color='g', label='Вектор v')
ax.quiver(0, 0, 0, normal[0], normal[1], normal[2], color='b', label='Вектор normal')

# Налаштування вісей
ax.set_xlim([-2, 4])
ax.set_ylim([-3, 3])
ax.set_zlim([-1, 4])
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.legend()

# Відображення графіку
plt.show()
```

#Мішаний добуток для визначення об'єму

![picture](https://drive.google.com/uc?id=1I_Y52s_WGUuX8Yy3bb2FhHp6PcuA6umk)




```python
import numpy as np

def parallelepiped_volume(a, b, c):
    # Обчислення мішаного добутку
    #mixed_dot_product = np.dot(a, np.cross(b, c))
    mixed_dot_product = a @ np.cross(b, c)

    # Обчислення об'єму паралелепіпеда
    volume = abs(mixed_dot_product)

    return volume

# Приклад використання
a = np.array([2, -2, -3])
b = np.array([4, 0, 6])
c = np.array([-7, -7, 1])

volume = parallelepiped_volume(a, b, c)
print("Об'єм паралелепіпеда:", volume)
```

#Визначення лінійної залежності засобами Numpy

```python
import numpy as np

def are_vectors_linearly_independent(vectors):
    # Створення розширеної матриці з векторів
    matrix = np.array(vectors).T

    # Ранг матриці
    rank_matrix = np.linalg.matrix_rank(matrix)

    # Кількість векторів
    num_vectors = len(vectors)

    # Вектори лінійно незалежні, якщо ранг матриці рівний кількості векторів
    return rank_matrix == num_vectors

# Приклад використання
#vectors1 = np.array([1, 2, 3])
#vectors2 = np.array([-2, 1, -1])
#vectors3 = np.array([3, 2, -1])

# Приклад використання
vectors1 = np.array([1, 2, -3])
vectors2 = np.array([-1, 2, 4])
vectors3 = np.array([1, 6, -2])

# Перевірка лінійної незалежності векторів
result = are_vectors_linearly_independent([vectors1, vectors2, vectors3])

if result:
    print("Вектори лінійно незалежні.")
else:
    print("Вектори лінійно залежні.")
```

##Робота з word embeddings ( словами)

Пам’ятайте, що розуміння даних є одним із найважливіших кроків у Data Science. Вбудовані слова є результатом процесів машинного навчання та будуть частиною вхідних даних для подальших процесів. Вбудовані слова – це багатовимірні масиви, як правило, із сотнями атрибутів, які створюють проблеми для їх інтерпретації.

У цьому зошиті ми візуально перевіримо вбудовування деяких слів за допомогою пари атрибутів. Необроблені атрибути не є найкращим варіантом для створення таких діаграм, але дозволять нам проілюструвати механічну частину в Python.

##Прогнозування столиць

Тепер, застосовуючи векторну різницю та додавання, можна створити векторне представлення для нового слова. Наприклад, ми можемо сказати, що векторна різниця між «Францією» та «Парижем» представляє концепцію "столиця".

Можна рухатися від міста Мадрид у напрямку поняття "столиця" та отримати щось близьке до відповідної країни, для якої Мадрид є столицею.

```python
capital = vec('France') - vec('Paris')
country = vec('Madrid') + capital

print(country[0:5]) # Print the first 5 values of the vector
```

```python
diff = country - vec('Spain')
print(diff[0:10])
```

Отже, ми повинні шукати найближчі слова у наборі даних, які відповідають країні-кандидату. Якщо вектори слів побудовані належним чином, найбільш схожим словом повинно бути «Spain». Визначимо функцію, яка допоможе нам це зробити. Ми будемо зберігати вектори слів як DataFrame, що полегшить операції пошуку на основі числових векторів.

```python
# Create a dataframe out of the dictionary embedding. This facilitate the algebraic operations
keys = word_embeddings.keys()
data = []
for key in keys:
    data.append(word_embeddings[key])

embedding = pd.DataFrame(data=data, index=keys)
# Define a function to find the closest word to a vector:
def find_closest_word(v, k = 1):
    # Calculate the vector difference from each word to the input vector
    diff = embedding.values - v
    # Get the squared L2 norm of each difference vector.
    # It means the squared euclidean distance from each word to the input vector
    delta = np.sum(diff * diff, axis=1)
    # Find the index of the minimun distance in the array
    i = np.argmin(delta)
    # Return the row name for this item
    return embedding.iloc[i].name
```

```python
find_closest_word(country)
```

```python
find_closest_word(vec('Italy') - vec('Rome') + vec('Madrid'))
```

```python
print(find_closest_word(vec('Berlin') + capital))
print(find_closest_word(vec('Beijing') + capital))
```
